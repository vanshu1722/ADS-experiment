**Experiment - 2 : Apply data cleaning techniques (e.g. Data Imputation)**

```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#load the data
data=pd.read_csv("/content/Tamilnadu agriculture yield data - Tamilnadu agriculture yield data.csv")
data.head()   #initial dataset
```

```python
#check for null values
print(data.isnull().sum())
```

```python
#data dimensions
print(data.shape)
```

**Most Frequent Value**

```python
from sklearn.impute import SimpleImputer
frequent_val=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
```

```python
#replacing all of the missing data with the most frequent value
data["State_Name"]=frequent_val.fit_transform(data[["State_Name"]])
data["District_Name"]=frequent_val.fit_transform(data[["District_Name"]])
data["Crop_Year"]=frequent_val.fit_transform(data[["Crop_Year"]])
data["Season"]=frequent_val.fit_transform(data[["Season"]])
data["Crop"]=frequent_val.fit_transform(data[["Crop"]])
data["Area"]=frequent_val.fit_transform(data[["Area"]])
data["Production"]=frequent_val.fit_transform(data[["Production"]])
```

```python
#printing the preprocessed dataset
data
```

```python
print(data.isna().sum())
```

**Using KNN Imputer**

```python
#load the data
data=pd.read_csv("/content/Tamilnadu agriculture yield data - Tamilnadu agriculture yield data.csv")
data.head()   #initial dataset
```

```python
print(data.isna().sum())
```

```python
#handle categorical data
new_data=pd.get_dummies(data,drop_first=True)
new_data #121 columns after handling categorical values
```

```python
#use KNN Imputer
from sklearn.impute import KNNImputer
knn=KNNImputer(n_neighbors=2)
imputed_data=knn.fit_transform(new_data)
final_data=pd.DataFrame(imputed_data)
```

**Mean or Median value**

```python
data=pd.read_csv("/content/Tamilnadu agriculture yield data - Tamilnadu agriculture yield data.csv")
data.head()
```

```python
#initialize simple imputer for median
med=SimpleImputer(missing_values=np.nan,strategy='median')

#apply frequent values on categorical columns
data["State_Name"]=frequent_val.fit_transform(data[["State_Name"]])
data["District_Name"]=frequent_val.fit_transform(data[["District_Name"]])
data["Crop_Year"]=frequent_val.fit_transform(data[["Crop_Year"]])
data["Season"]=frequent_val.fit_transform(data[["Season"]])
data["Crop"]=frequent_val.fit_transform(data[["Crop"]])

#apply median values on numerical columns
data["Area"]=med.fit_transform(data[["Area"]])
data["Production"]=med.fit_transform(data[["Production"]])
```

```python
#imputed data now
data
```

```python
print(data.isnull().sum())
```

**Fixed Value**

```python
data=pd.read_csv("/content/Tamilnadu agriculture yield data - Tamilnadu agriculture yield data.csv")
data.head()
```

```python
print(data.isna().sum())
```

```python
#apply fill values to columns using pandas
data["State_Name"].fillna("Tamil Nadu",inplace=True)
data["District_Name"].fillna("COIMBATORE",inplace=True)
data["Crop_Year"].fillna("2000",inplace=True)
data["Season"].fillna("Kharif",inplace=True)
data["Crop"].fillna("Maize",inplace=True)
data["Area"].fillna("5500",inplace=True)
data["Production"].fillna("12000",inplace=True)

#print the imputed dataset
data.head()
```

```python
print(data.isna().sum())
```

**Replace missing values with min/max value**

```python
data=pd.read_csv("/content/Tamilnadu agriculture yield data - Tamilnadu agriculture yield data.csv")
data.tail()
```

```python
print(data.isna().sum())
```

```python
#apply max/min values
data["State_Name"].fillna("Tamil Nadu",inplace=True)
data["District_Name"].fillna("CHENNAI",inplace=True)
data["Crop_Year"].fillna("2000",inplace=True)
data["Season"].fillna("Kharif",inplace=True)
data["Crop"].fillna("Maize",inplace=True)

#fill with max values
max_area=data["Area"].max()
data["Area"].fillna(min_area,inplace=True)

max_prod=data["Production"].max()
data["Production"].fillna(max_prod,inplace=True)
print(data.isna().sum())
```
