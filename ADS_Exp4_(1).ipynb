```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Part 1 - Supervised Learning**

```python
#https://www.kaggle.com/datasets/nelgiriyewithana/apple-quality
data=pd.read_csv("/content/apple_quality.csv")
data.drop(["A_id"],axis='columns',inplace=True)
data
```

```python
print(data.isna().sum())
```

```python
data.info()
```

```python
data.describe()
```

```python
features=data.drop(["Quality"],axis='columns')
target=data["Quality"]
```

```python
from sklearn.preprocessing import MinMaxScaler,LabelEncoder

mms=MinMaxScaler()
scaled_features=mms.fit_transform(features)
final_features=pd.DataFrame(scaled_features,columns=["Size","Weight","Sweetness","Crunchiness","Juiciness","Ripeness","Acidity"])
final_features

le=LabelEncoder()
data["target"]=le.fit_transform(target)
data["target"].unique()

new_target=data["target"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_features,new_target,test_size=0.2,random_state=0)
model=KNeighborsClassifier(n_neighbors=63,metric='euclidean')
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import *
y_pred=model.predict(x_test)

accuracy=accuracy_score(y_test,y_pred)
print("Accuracy of this model is: ",round(accuracy*100,2),"%")

precision=precision_score(y_test,y_pred)
print("Precision of this model is= ",round(precision,2))

recall=recall_score(y_test,y_pred)
print("Recall score of this model is= ",round(recall,2))

f1=f1_score(y_test,y_pred)
print("F1 Score of this model is= ",round(f1,2))

roc_auc=roc_auc_score(y_test,y_pred)
print("ROC AUC of this model is= ",round(roc_auc,2))

mae=mean_absolute_error(y_test,y_pred)
print("Mean absolute error of this model is= ",round(mae,2))

mse=mean_squared_error(y_test,y_pred)
print("Mean squared error of this model is= ",round(mse,2))
```

**Part 2 - Unsupervised Learning**

```python
data=pd.read_csv("/content/cleaned_data.csv") #import the preprocessed dataset
data.head()
```

```python
features=data[["Area","Production"]]
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=3)
data["Cluster"]=model.fit_predict(features)
```

```python
#metrics for this model
sil_score=silhouette_score(features,data["Cluster"])
print("Silhouette score for this model is: ",round(sil_score,3))
ch_score=calinski_harabasz_score(features,data["Cluster"])
print("Calinski Harabasz score for this model is: ",round(ch_score,2))
db_score=davies_bouldin_score(features,data["Cluster"])
print("Davies Bouldin score for this model is: ",round(db_score,2))
```
